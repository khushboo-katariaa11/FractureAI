```python
import os
import shutil
import random

src_train_dir = '/content/mura_v11/MURA-v1.1/train/XR_WRIST'
src_valid_dir = '/content/mura_v11/MURA-v1.1/valid/XR_WRIST'

out_root = 'dataset_wrist'
splits = ['train', 'valid', 'test']
labels_map = {'positive': 'fracture', 'negative': 'normal'}  # Map folder keyword to label

all_studies = []

def gather_images(src_dir):
    for patient_id in os.listdir(src_dir):
        patient_path = os.path.join(src_dir, patient_id)
        if not os.path.isdir(patient_path):
            continue
        for study_folder in os.listdir(patient_path):
            study_path = os.path.join(patient_path, study_folder)
            if not os.path.isdir(study_path):
                continue
            label_flag = None
            if 'positive' in study_folder:
                label_flag = 'fracture'
            elif 'negative' in study_folder:
                label_flag = 'normal'
            else:
                continue  # Skip unknown
            for img_file in os.listdir(study_path):
                if not (img_file.lower().endswith('.png') or img_file.lower().endswith('.jpg')):
                    continue
                img_path = os.path.join(study_path, img_file)
                all_studies.append({
                    'patient': patient_id,
                    'study_folder': study_folder,
                    'img_file': img_file,
                    'label': label_flag,
                    'img_path': img_path
                })

# Gather images from train and valid
gather_images(src_train_dir)
gather_images(src_valid_dir)

# Group patients by label
from collections import defaultdict
patients_by_label = defaultdict(set)
for s in all_studies:
    patients_by_label[s['label']].add(s['patient'])

for label in patients_by_label:
    patients_by_label[label] = list(patients_by_label[label])

# Split by patient for each label
random.seed(42)
split_patient_ids = {split: {'fracture': [], 'normal': []} for split in splits}
split_ratios = {'train': 0.7, 'valid': 0.15, 'test': 0.15}

for label in ['fracture', 'normal']:
    pats = patients_by_label[label]
    random.shuffle(pats)
    n = len(pats)
    split_patient_ids['train'][label] = pats[:int(n*split_ratios['train'])]
    split_patient_ids['valid'][label] = pats[int(n*split_ratios['train']):int(n*(split_ratios['train']+split_ratios['valid']))]
    split_patient_ids['test'][label] = pats[int(n*(split_ratios['train']+split_ratios['valid'])):]

# Create folder structure
for split in splits:
    for label in ['fracture', 'normal']:
        os.makedirs(os.path.join(out_root, split, label), exist_ok=True)

# Copy images accordingly
for s in all_studies:
    for split in splits:
        if s['patient'] in split_patient_ids[split][s['label']]:
            dest_dir = os.path.join(out_root, split, s['label'])
            dest_path = os.path.join(dest_dir, f"{s['patient']}_{s['study_folder']}_{s['img_file']}")
            shutil.copy2(s['img_path'], dest_path)

# Print summary
for split in splits:
    for label in ['fracture', 'normal']:
        path = os.path.join(out_root, split, label)
        print(f"{path}: {len(os.listdir(path))} images")
```

```python
from torchvision import transforms

# For training
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),                         # Standard input size for most ImageNet CNNs
    transforms.RandomHorizontalFlip(),                     # Augmentation for train only
    transforms.RandomRotation(10),                         # Small rotation for anatomical variation
    transforms.ColorJitter(brightness=0.1, contrast=0.1),  # Subtle pixel-level jitter
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],            # ImageNet stats
                         [0.229, 0.224, 0.225]),
])

# For validation and test
val_test_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])
```

```python
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader

# Paths to your organized dataset folders
train_dir = 'dataset_wrist/train'
valid_dir = 'dataset_wrist/valid'
test_dir = 'dataset_wrist/test'

# Custom filter for hidden/corrupt files (optional but smart)
import os
def is_valid_image(path):
    if os.path.basename(path).startswith("._"):
        return False
    if os.path.getsize(path) == 0:
        return False
    return True

# Setup datasets (ImageFolder expects: /class_name/image.png structure)
train_dataset = ImageFolder(train_dir, transform=train_transforms, is_valid_file=is_valid_image)
val_dataset = ImageFolder(valid_dir, transform=val_test_transforms, is_valid_file=is_valid_image)
test_dataset = ImageFolder(test_dir, transform=val_test_transforms, is_valid_file=is_valid_image)

# Dataloaders
BATCH_SIZE = 32

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=4)

print("Loaded samples:")
print("Train:", len(train_dataset), " Valid:", len(val_dataset), " Test:", len(test_dataset))
print("Classes:", train_dataset.classes)
```

```python
import torch
import torch.nn as nn
import torchvision.models as models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load DenseNet121 pretrained on ImageNet
model = models.densenet121(weights=models.DenseNet121_Weights.IMAGENET1K_V1)
num_features = model.classifier.in_features
model.classifier = nn.Linear(num_features, 2)  # 2 classes: fracture, normal
model = model.to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-4)
# Optional: Adaptive learning rate on plateau
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2, verbose=True)
```

```python
def is_valid_image(path):
    fname = os.path.basename(path)
    # Exclude any file containing '._'
    if '._' in fname or fname.startswith('.'):
        return False
    if os.path.getsize(path) == 0:
        return False
    valid_ext = ['.png', '.jpg', '.jpeg']
    if not any(fname.lower().endswith(e) for e in valid_ext):
        return False
    return True

```

```python
NUM_EPOCHS = 10
best_val_acc = 0.0

for epoch in range(NUM_EPOCHS):
    print(f"\nEpoch {epoch+1}/{NUM_EPOCHS}")
    print("-" * 20)

    # --- Training ---
    model.train()
    running_loss, running_corrects = 0.0, 0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, preds = torch.max(outputs, 1)
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)

    train_loss = running_loss / len(train_loader.dataset)
    train_acc = running_corrects.double() / len(train_loader.dataset)
    print(f"Train Loss: {train_loss:.4f}, Acc: {train_acc:.4f}")

    # --- Validation ---
    model.eval()
    val_loss, val_corrects = 0.0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            _, preds = torch.max(outputs, 1)
            val_loss += loss.item() * inputs.size(0)
            val_corrects += torch.sum(preds == labels.data)

    val_loss = val_loss / len(val_loader.dataset)
    val_acc = val_corrects.double() / len(val_loader.dataset)
    print(f"Validation Loss: {val_loss:.4f}, Acc: {val_acc:.4f}")

    scheduler.step(val_acc)

    # --- Save best model ---
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), "best_densenet121.pth")
        print("✅ Best model saved")

print(f"\nBest validation accuracy: {best_val_acc:.4f}")
```

```python
import torch
from sklearn.metrics import accuracy_score

def get_accuracy(model, data_loader, device):
    model.eval()
    preds = []
    labels = []
    with torch.no_grad():
        for inputs, targets in data_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            _, batch_preds = torch.max(outputs, 1)
            preds.extend(batch_preds.cpu().numpy())
            labels.extend(targets.cpu().numpy())
    return accuracy_score(labels, preds)

# Get accuracies
train_accuracy = get_accuracy(model, train_loader, device)
test_accuracy = get_accuracy(model, test_loader, device)

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy:     {test_accuracy:.4f}")
```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Image as RLImage, Table, TableStyle
from reportlab.lib.pagesizes import letter
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib import colors
from PIL import Image
import numpy as np
import os
from datetime import datetime

def save_pdf_report(
    idx,
    model,
    test_dataset,
    device,
    hospital="AI Imaginary Hospital",
    doctor="Dr. AI Wonder",
    exam_desc="Wrist X-ray for fracture detection.",
    patient_meta=None,
    save_pdf=True,
    report_dir="reports_pdf"
):
    # Set up basic info
    img_path, true_label_idx = test_dataset.samples[idx]
    filename = os.path.basename(img_path)
    name, sex, age, pat_id, description = "N/A", "N/A", "N/A", "N/A", exam_desc
    if patient_meta and filename in patient_meta:
        md = patient_meta[filename]
        name = md.get('name', name)
        sex = md.get('sex', sex)
        age = md.get('age', age)
        pat_id = md.get('id', pat_id)
        description = md.get('desc', description)

    # Model input and prediction
    from torchvision import transforms
    import torch
    image_pil = Image.open(img_path).convert('RGB')
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    input_tensor = transform(image_pil).unsqueeze(0).to(device)
    model.eval()
    with torch.no_grad():
        outputs = model(input_tensor)
        probs = torch.softmax(outputs, 1)
        _, predicted_label_idx = torch.max(probs, 1)
    class_names = test_dataset.classes
    pred_label = class_names[predicted_label_idx.item()]
    true_label = class_names[true_label_idx]
    pred_prob = probs[0, predicted_label_idx].item()

    # Grad-CAM image
    from pytorch_grad_cam import GradCAM
    from pytorch_grad_cam.utils.image import show_cam_on_image
    target_layers = [model.features[-1]]
    cam = GradCAM(model=model, target_layers=target_layers)
    grayscale_cam = cam(input_tensor=input_tensor)[0]
    img_norm = np.array(image_pil.resize((224, 224))) / 255.0
    cam_image = show_cam_on_image(img_norm, grayscale_cam, use_rgb=True)

    # Save images temporarily for PDF embedding
    orig_img_path = f"tmp_{filename}_orig.png"
    cam_img_path = f"tmp_{filename}_cam.png"
    image_pil.resize((224, 224)).save(orig_img_path)
    Image.fromarray(cam_image).save(cam_img_path)

    # PDF Report
    os.makedirs(report_dir, exist_ok=True)
    dt_str = datetime.now().strftime('%Y%m%d_%H%M%S')
    pdf_filename = f"report_{filename}_{dt_str}.pdf"
    pdf_path = os.path.join(report_dir, pdf_filename)

    doc = SimpleDocTemplate(pdf_path, pagesize=letter)
    styles = getSampleStyleSheet()
    flow = []

    # Heading & key metadata
    flow.append(Paragraph(f"<b>{hospital}</b>", styles['Title']))
    flow.append(Spacer(1, 16))
    flow.append(Paragraph("<b>Radiologist:</b> " + doctor, styles['Normal']))
    flow.append(Paragraph("<b>Date:</b> " + datetime.now().strftime('%Y-%m-%d %H:%M'), styles['Normal']))
    data = [
        ["Patient ID", pat_id, "Name", name],
        ["Sex", sex, "Age", str(age)],
        ["Exam", "Wrist X-ray", "Filename", filename],
    ]
    t = Table(data)
    t.setStyle(TableStyle([('BACKGROUND', (0, 0), (-1, 0), colors.grey),('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),('ALIGN',(0,0),(-1,-1),'LEFT'),('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),('BOTTOMPADDING', (0,0), (-1,0), 12),('BACKGROUND',(0,1),(-1,-1),colors.beige)]))
    flow.append(t)
    flow.append(Spacer(1, 16))

    # Clinical description
    flow.append(Paragraph("<b>Description:</b>", styles['Normal']))
    flow.append(Paragraph(description, styles['Normal']))
    flow.append(Spacer(1, 12))

    # AI assessment
    flow.append(Paragraph("<b>AI Model Fracture Assessment:</b>", styles['Normal']))
    flow.append(Paragraph(f"<b>Prediction:</b> {pred_label.upper()} ({pred_prob:.2%} confidence)", styles['Normal']))
    flow.append(Paragraph(f"<b>Ground Truth:</b> {true_label.upper()}", styles['Normal']))
    flow.append(Paragraph(f"<b>Correct:</b> {'YES ✅' if pred_label==true_label else 'NO ❌'}", styles['Normal']))
    flow.append(Spacer(1, 12))

    # Summary findings
    flow.append(Paragraph("<b>Findings & Summary:</b>", styles['Normal']))
    summ = f"""
    <ul>
    <li>({pred_label.upper()}) detected by AI fracture model.</li>
    <li>If FRACTURE, recommend orthopedic consult.</li>
    <li>If NORMAL, but clinical suspicion strong, follow-up imaging suggested.</li>
    </ul>
    """
    flow.append(Paragraph(summ, styles['Normal']))
    flow.append(Spacer(1, 12))

    # Images
    flow.append(Paragraph("<b>Original Image</b>", styles['Normal']))
    flow.append(RLImage(orig_img_path, width=200, height=200))
    flow.append(Spacer(1, 8))
    flow.append(Paragraph("<b>Grad-CAM Activation</b>", styles['Normal']))
    flow.append(RLImage(cam_img_path, width=200, height=200))
    flow.append(Spacer(1, 12))

    flow.append(Paragraph("<i>End of report (auto-generated by AI FractureNet)</i>", styles['Normal']))

    doc.build(flow)
    print(f"PDF report saved to: {pdf_path}")

    # Clean up temp images
    os.remove(orig_img_path)
    os.remove(cam_img_path)

    return pdf_path
```
